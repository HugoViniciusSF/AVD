Media do meta_score de cada genero por plataforma

```python
# Exclua os valores NaN das colunas 'Genre', 'Platform', 'Critic_Score' e 'User_Score'
gData_sem_nan = gData.dropna(subset=['Genre', 'Platform', 'Critic_Score', 'User_Score']).copy()

# Converta o User_Score para o tipo num√©rico (float) e multiplique por 10
gData_sem_nan.loc[:, 'User_Score'] = pd.to_numeric(gData_sem_nan['User_Score'], errors='coerce') * 10

# Calcule o meta_score como a m√©dia entre Critic_Score e User_Score
gData_sem_nan.loc[:, 'Meta_Score'] = (gData_sem_nan['Critic_Score'] + gData_sem_nan['User_Score']) / 2

# Agrupe os dados por g√™nero e plataforma
grupos_por_genero_plataforma = gData_sem_nan.groupby(['Genre', 'Platform'])

# Calcule a m√©dia do Meta_Score para cada grupo
media_por_genero_plataforma = grupos_por_genero_plataforma['Meta_Score'].mean()

print(media_por_genero_plataforma.round)
```

Aqui eu sei quanto cada publicadora participou na media do do meta_score do genero por plataforma.

```python
def calcular_participacoes(dados, genero, plataforma):
    # Filtrar os dados apenas para o g√™nero e plataforma espec√≠ficos, removendo valores NaN
    dados_genero_plataforma = dados[(dados['Genre'] == genero) & (dados['Platform'] == plataforma)].dropna(subset=['Meta_Score'])

    # Obter a lista √∫nica de publicadoras que lan√ßaram jogos nessa plataforma e g√™nero
    publicadoras = dados_genero_plataforma['Publisher'].unique()

    # Calcular a soma total do Meta_Score para o g√™nero e plataforma espec√≠ficos
    soma_total_meta_score_plataforma = dados_genero_plataforma['Meta_Score'].sum()

    # Criar um dicion√°rio para armazenar as participa√ß√µes percentuais de cada publicadora
    participacao_publicadoras = {}

    # Iterar sobre cada publicadora
    for publicadora in publicadoras:
        # Filtrar os dados para os jogos dessa publicadora, removendo valores NaN
        jogos_publicadora = dados_genero_plataforma[dados_genero_plataforma['Publisher'] == publicadora].dropna(subset=['Meta_Score'])

        # Calcular a soma do Meta_Score apenas para os jogos dessa publicadora
        soma_meta_score_publicadora = jogos_publicadora['Meta_Score'].sum()

        # Calcular a participa√ß√£o percentual da publicadora no valor total do Meta_Score para o g√™nero e plataforma espec√≠ficos
        participacao_publicadora = (soma_meta_score_publicadora / soma_total_meta_score_plataforma) * 100

        # Armazenar a participa√ß√£o percentual no dicion√°rio
        participacao_publicadoras[publicadora] = participacao_publicadora

    return participacao_publicadoras


def imprimir_participacoes(participacao_publicadoras, genero, plataforma):
    # Imprimir as participa√ß√µes percentuais de cada publicadora
    for publicadora, participacao in participacao_publicadoras.items():
        print(f"A participa√ß√£o da publicadora '{publicadora}' no valor total do Meta_Score para o g√™nero '{genero}' na plataforma '{plataforma}' √© de {participacao:.2f}%.")

    # Calcular e imprimir a soma das participa√ß√µes
    soma_participacoes = sum(participacao_publicadoras.values())
    print(f"\nA soma das participa√ß√µes √© de {soma_participacoes:.2f}%.")


# Escolha a plataforma e o g√™nero espec√≠ficos
plataforma_especifica = 'PS4'
genero_especifico = 'Action'

# Calcular as participa√ß√µes
participacao_publicadoras = calcular_participacoes(gData_sem_nan, genero_especifico, plataforma_especifica)

# Imprimir as participa√ß√µes
imprimir_participacoes(participacao_publicadoras, genero_especifico, plataforma_especifica)
```

```python
def calcular_participacoes(dados, genero, plataforma):
    # Filtrar os dados apenas para o g√™nero e plataforma espec√≠ficos, removendo valores NaN
    dados_genero_plataforma = dados[(dados['Genre'] == genero) & (dados['Platform'] == plataforma)].dropna(subset=['Meta_Score'])

    # Obter a lista √∫nica de publicadoras que lan√ßaram jogos nessa plataforma e g√™nero
    publicadoras = dados_genero_plataforma['Publisher'].unique()

    # Calcular a soma total do Meta_Score para o g√™nero e plataforma espec√≠ficos
    soma_total_meta_score_plataforma = dados_genero_plataforma['Meta_Score'].sum()

    # Criar um dicion√°rio para armazenar as participa√ß√µes percentuais de cada publicadora
    participacao_publicadoras = {}

    # Iterar sobre cada publicadora
    for publicadora in publicadoras:
        # Filtrar os dados para os jogos dessa publicadora, removendo valores NaN
        jogos_publicadora = dados_genero_plataforma[dados_genero_plataforma['Publisher'] == publicadora].dropna(subset=['Meta_Score'])

        # Calcular a soma do Meta_Score apenas para os jogos dessa publicadora, se houver
        if not jogos_publicadora.empty:
            soma_meta_score_publicadora = jogos_publicadora['Meta_Score'].sum()

            # Calcular a participa√ß√£o percentual da publicadora no valor total do Meta_Score para o g√™nero e plataforma espec√≠ficos
            participacao_publicadora = (soma_meta_score_publicadora / soma_total_meta_score_plataforma) * 100

            # Armazenar a participa√ß√£o percentual no dicion√°rio
            participacao_publicadoras[publicadora] = participacao_publicadora

    return participacao_publicadoras


def imprimir_participacoes(participacao_publicadoras, genero, plataforma):
    # Imprimir as participa√ß√µes percentuais de cada publicadora
    for publicadora, participacao in participacao_publicadoras.items():
        print(f"A participa√ß√£o da publicadora '{publicadora}' no valor total do Meta_Score para o g√™nero '{genero}' na plataforma '{plataforma}' √© de {participacao:.2f}%.")

    # Calcular e imprimir a soma das participa√ß√µes
    soma_participacoes = sum(participacao_publicadoras.values())
    print(f"\nA soma das participa√ß√µes √© de {soma_participacoes:.2f}%.")


# Escolha a plataforma e o g√™nero espec√≠ficos
plataforma_especifica = '3DS'
genero_especifico = 'Action'

# Calcular as participa√ß√µes
participacao_publicadoras = calcular_participacoes(gData_sem_nan, genero_especifico, plataforma_especifica)

# Imprimir as participa√ß√µes
imprimir_participacoes(participacao_publicadoras, genero_especifico, plataforma_especifica)
```
